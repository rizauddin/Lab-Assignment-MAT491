Parametric Equations
-------------------

**Example:** Graph the curve $x = \sin{t}$, $y = \cos{t}$, $0 \le t \le \pi$.

Read Scipy Lecture Notes pg. 45 for creating arrays and pg. 83 for plotting. (Scipy Lecture Notes - http://www.scipy-lectures.org/)

```python
# define array from 0 to pi with interval size 0.1
t = np.arange(0, np.pi, 0.1)

#define x and y
x = np.sin(t)
y = np.cos(t)

# plot the graph
plt.plot(x, y)
plt.show()
```

Polar Coordinates
----------------

**Example:** Sketch the curve $r = 1 + \sin{\theta}$.

```python
# define array from 0 to pi 100 elements
theta = np.linspace(0, 2*np.pi, 100)  

# define $r = 1 + \sin{\theta}$
r = 1 + np.sin(theta)

# add subplot in polar coordinates 
ax = plt.subplot(111, polar=True)

# plot the graph
ax.plot(theta, r)
plt.show()
```
